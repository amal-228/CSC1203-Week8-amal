```python
import os

lab4_data_folder = 'lab4_data'
subfolders = ['input', 'output', 'logs']

for subfolder in subfolders:
    subfolder_path = os.path.join(lab4_data_folder, subfolder)
    if not os.path.exists(subfolder_path):
        os.makedirs(subfolder_path)
```

```python
import os

lab4_data_folder = 'lab4_data'

for item in os.listdir(lab4_data_folder):
    item_path = os.path.join(lab4_data_folder, item)
    if os.path.isdir(item_path):
        print(f'{item} is a directory.')
    else:
        print(f'{item} is a file.')
```

```python
import os

lab4_data_folder = 'lab4_data'

logs_folder_path = os.path.join(lab4_data_folder, 'logs')
history_folder_path = os.path.join(lab4_data_folder, 'history')

if os.path.exists(logs_folder_path) and not os.path.exists(history_folder_path):
    os.rename(logs_folder_path, history_folder_path)
    print("Renamed 'logs' to 'history'")
else:
    print("Rename skipped (either logs doesn't exist or history already exists)")
```

```python
import os

lab4_data_folder = 'lab4_data'
input_folder = os.path.join(lab4_data_folder, 'input')
output_folder = os.path.join(lab4_data_folder, 'output')

notes_file_path = os.path.join(input_folder, 'notes.txt')
summary_file_path = os.path.join(output_folder, 'summary.txt')

with open(notes_file_path, 'r') as file:
    lines = file.readlines()

with open(summary_file_path, 'w') as file:
    file.writelines(lines)
```

```python
from pyfiglet import figlet_format
import os

lab4_data_folder = 'lab4_data'
output_folder = os.path.join(lab4_data_folder, 'output')

ascii_art = figlet_format("Amal")

print(ascii_art)

ascii_file_path = os.path.join(output_folder, 'ascii.txt')

with open(ascii_file_path, 'w') as file:
    file.write(ascii_art)
```
